Question 1: What is Simple Linear Regression?

Answer: "Simple Linear Regression" is a statistical technique used to understand the relationship between two variables. In this method, one variable is independent (input) and the other variable is dependent (output). It shows how a change in one variable affects the other.

The main purpose of simple linear regression is to predict the value of the dependent variable based on the independent variable. It does this by fitting a straight line that best represents the data points. This straight line shows the overall trend between the two variables.

For example, if we want to predict a student’s marks based on the number of hours studied, the study hours are the independent variable and the marks are the dependent variable. As the study hours increase, the marks generally increase, and this relationship can be shown using simple linear regression.

The relationship between the variables is represented by the equation "y = mx + c", where 'y' is the dependent variable, 'x' is the independent variable, 'm' is the slope of the line, and 'c' is the intercept. Overall, simple linear regression is widely used in data analysis and prediction because it is easy to understand and apply.

Question 2: What are the key assumptions of Simple Linear Regression?

Answer: Key Assumptions of Simple Linear Regression

Linearity: The dependent and independent variables change in a straight line.

1) Independent observations: Every data point is unrelated to the others and stands alone.

2) Error normality: The residuals have a normal distribution.

3) Zero error mean: The average error value is zero. The spread of errors is constant across all values of the independent variable (homoscedasticity).

4) No autocorrelation: There is no correlation between the residuals
5) Precise measurement of X: There is no inaccuracy in the measurement of the independent variable.
6) No extreme outliers: The regression line is not significantly impacted by any odd numbers.
7) Continuous dependent variable: The dependent variable is both continuous and numerical.
8) Enough variety in X: Changes in Y can be explained by the independent variable.

Question 3: What is heteroscedasticity, and why is it important to address in regression models?

Answer: Heteroscedasticity

When the variance of the error terms in a regression model is not constant for all values of the independent variable, this is known as heteroscedasticity.  This indicates that when the independent variable's value rises or falls, the residuals' distribution varies.  One of the fundamental tenets of linear regression is violated by heteroscedasticity, which frequently occurs in real-world data, particularly in economic and commercial datasets.

Importance of Heteroscedasticity

1) It contributes to precise standard error estimation.

2) It increases the reliability of hypothesis tests (t-test and F-test).

3) It results in accurate regression coefficient confidence intervals.

4) It keeps variable significance from being misinterpreted.

5) It enhances the regression model's overall validity and reliability.

6) It aids in improving model-based forecasts and choices.


Question 4: What is Multiple Linear Regression?

Answer: Multiple Linear Regression

A statistical method for determining the association between one dependent variable and two or more independent variables is called multiple linear regression.  numerous linear regression takes into account numerous factors simultaneously to explain or predict the result, in contrast to simple linear regression, which employs only one predictor.

 This approach modifies the dependent variable according to the total impact of all independent variables.  For instance, a house's age, location, area, and number of rooms can all be used to estimate its price.  When other factors are held constant, multiple linear regression can be used to determine how each independent variable affects the dependent variable.

y = b₀ + b₁x₁ + b₂x₂ + … + bₙxₙ

Where:

y = dependent variable

x₁, x₂, …, xₙ = independent variables

b₀ = intercept (starting value)

b₁, b₂, …, bₙ = coefficients that show the effect of each independent variable on y


Question 5: What is polynomial regression, and how does it differ from linear regression?

Answer: When there is a non-linear relationship between the independent and dependent variables, polynomial regression analysis is utilised.  Polynomial regression uses powers of the independent variable to model the data as a curved line, in contrast to simple linear regression, which assumes a straight-line relationship.  This makes it possible for the model to identify more intricate patterns in the data that a straight line is unable to depict.

 The following is the generic formula for polynomial regression:
 y = b₀ + b₁x + b₂x² + b₃x³ + … + bₙxⁿ

Where:

y = dependent variable

x = independent variable

b₀ = intercept

b₁, b₂, …, bₙ = coefficients for each power of x

n = degree of the polynomial

Key Differences from Linear Regression:

1) Relationship Type: Polynomial regression models curves, whereas linear regression assumes a straight-line relationship.

2) Flexibility: Unlike linear regression, polynomial regression can accommodate more intricate data patterns.

3) Equation Form: Higher powers of x (x² x³, etc.) are included in polynomial regression, while y = b₀ + b₁x is used in linear regression.

4) Use Case: Polynomial regression is employed when data clearly curves or bends, while linear regression is effective when data is roughly linear.




```python
#Question 6: Implement a Python program to fit a Simple Linear Regression model to the following sample data:
#  X = [1, 2, 3, 4, 5]
# Y = [2.1, 4.3, 6.1, 7.9, 10.2]
# Plot the regression line over the data points.

# Step 1: Import required libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Step 2: Define the data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)  # Independent variable
Y = np.array([2.1, 4.3, 6.1, 7.9, 10.2])     # Dependent variable

# Step 3: Create and fit the Linear Regression model
model = LinearRegression()
model.fit(X, Y)

# Step 4: Predict Y values using the model
Y_pred = model.predict(X)

# Step 5: Plot the data points
plt.scatter(X, Y, color='blue', label='Data points')

# Step 6: Plot the regression line
plt.plot(X, Y_pred, color='red', label='Regression line')

# Step 7: Add labels and title
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Simple Linear Regression')
plt.legend()
plt.show()

# Step 8: Print the slope and intercept
print(f"Slope (m): {model.coef_[0]}")
print(f"Intercept (c): {model.intercept_}")
```

```python
#Question 7: Fit a Multiple Linear Regression model on this sample data:
# Area = [1200, 1500, 1800, 2000]
# Rooms = [2, 3, 3, 4]
# Price = [250000, 300000, 320000, 370000]
#Check for multicollinearity using VIF and report the results.
#(Include your Python code and output in the code box below.)

# Step 1: Import required libraries
import pandas as pd
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Step 2: Define the data
data = pd.DataFrame({
    'Area': [1200, 1500, 1800, 2000],
    'Rooms': [2, 3, 3, 4],
    'Price': [250000, 300000, 320000, 370000]
})

# Step 3: Define independent (X) and dependent (Y) variables
X = data[['Area', 'Rooms']]
Y = data['Price']

# Step 4: Fit the Multiple Linear Regression model
model = LinearRegression()
model.fit(X, Y)

# Step 5: Print coefficients and intercept
print("Coefficients:", model.coef_)
print("Intercept:", model.intercept_)

# Step 6: Check for multicollinearity using VIF
# Add constant term for VIF calculation
X_const = sm.add_constant(X)

# Calculate VIF for each feature
vif = pd.DataFrame()
vif['Feature'] = X_const.columns
vif['VIF'] = [variance_inflation_factor(X_const.values, i) for i in range(X_const.shape[1])]

print("\nVariance Inflation Factor (VIF):")
print(vif)

# # # Regression Equation: Price=103157.89+63.16×Area+34736.84×Rooms
# Each additional square foot of Area increases Price by ~63.16 units.
# Each additional Room increases Price by ~34,736.84 units (keeping Area constant).

#VIF Interpretation: Area VIF = 7.74

#Rooms VIF = 7.74
#Since VIF > 5, there is high multicollinearity.
#High multicollinearity means Area and Rooms are strongly correlated, making the coefficients less reliable.
```

```python
#Question 8: Implement polynomial regression on the following data:
# X = [1, 2, 3, 4, 5]
# Y = [2.2, 4.8, 7.5, 11.2, 14.7]
# Fit a 2nd-degree polynomial and plot the resulting curve.

# Step 1: Import required libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# Step 2: Define the data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)  # Independent variable
Y = np.array([2.2, 4.8, 7.5, 11.2, 14.7])     # Dependent variable

# Step 3: Transform X to include polynomial features (degree=2)
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

# Step 4: Fit Linear Regression on the transformed data
model = LinearRegression()
model.fit(X_poly, Y)

# Step 5: Predict Y values using the model
Y_pred = model.predict(X_poly)

# Step 6: Plot the original data points
plt.scatter(X, Y, color='blue', label='Data points')

# Step 7: Plot the polynomial regression curve
plt.plot(X, Y_pred, color='red', label='Polynomial curve (degree=2)')

# Step 8: Add labels and title
plt.xlabel('X')
plt.ylabel('Y')
plt.title('2nd Degree Polynomial Regression')
plt.legend()
plt.show()

# Step 9: Print the coefficients
print(f"Intercept: {model.intercept_}")
print(f"Coefficients: {model.coef_}")
```

```python
#Question 9: Create a residuals plot for a regression model trained on this data:
# X = [10, 20, 30, 40, 50]
# Y = [15, 35, 40, 50, 65]
# Assess heteroscedasticity by examining the spread of residuals

# Step 1: Import libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Step 2: Define the data
X = np.array([10, 20, 30, 40, 50]).reshape(-1, 1)
Y = np.array([15, 35, 40, 50, 65])

# Step 3: Fit Linear Regression model
model = LinearRegression()
model.fit(X, Y)

# Step 4: Predict Y values
Y_pred = model.predict(X)

# Step 5: Calculate residuals
residuals = Y - Y_pred

# Step 6: Create residuals plot
plt.scatter(X, residuals, color='blue')
plt.axhline(y=0, color='red', linestyle='--')
plt.xlabel('X')
plt.ylabel('Residuals (Y - Y_pred)')
plt.title('Residuals Plot')
plt.show()

# Step 7: Print residuals
print("Residuals:", residuals)


######Residuals are calculated as:Residual=Y−Y^
#where 𝑌^, is the predicted value from the regression model.

#The residuals plot shows residuals on the y-axis and X on the x-axis.

#The red dashed line at y=0 helps you see deviations from the model.
```

Question 10: Imagine you are a data scientist working for a real estate company. You need to predict house prices using features like area, number of rooms, and location. However, you detect heteroscedasticity and multicollinearity in your regression model. Explain the steps you would take to address these issues and ensure a robust model.

Answer: As a data scientist, identifying heteroscedasticity and multicollinearity in a regression model to forecast home prices suggests possible issues with the model's dependability.  I would take the following actions to resolve these problems and guarantee a reliable model:

1) Handling Heteroscedasticity: When residual variance is not constant, it can lead to incorrect statistical tests and predictions. To make it right:

   i) Change the dependent variable:  To stabilise variance, use Box-Cox, square root, or logarithmic transformations.  Consider using log(Price) for Price.

   ii) Weighted Least Squares (WLS): To lessen heteroscedasticity, give observations weights that are inversely proportionate to their variance
   
   iii) Examine residuals following transformation:  To verify that variance is more constant, plot residuals once more.

2) Addressing Multicollinearity: Multicollinearity happens when independent variables are highly correlated, making it difficult to interpret coefficients. To fix it:

   i) Remove or combine correlated features: For example, if area and number of rooms are highly correlated, consider keeping only one or combining them into a single feature.
   
   ii) Use dimensionality reduction: Techniques like Principal Component Analysis (PCA) can combine correlated variables into new uncorrelated components.
   
   iii) Regularization methods: Use Ridge or Lasso regression to penalize large coefficients and reduce the effect of multicollinearity.

3. Model Validation and Robustness

   i) Cross-validation: Use k-fold cross-validation to ensure the model generalizes well to unseen data.
   
   ii) Residual analysis: After corrections, always check residual plots to ensure homoscedasticity.
   
   iii) Feature engineering: Explore additional features like location scores or proximity to amenities to improve model accuracy.
